```python
# barplot
res=sns.barplot(mtcars['cyl'],mtcars['mpg'])
plt.show()
```

```python
res=sns.barplot(x='cyl',y='mpg',data=mtcars,color='blue')
plt.show
```

```python
res=sns.barplot(mtcars['gear'],mtcars['hp'],palette='rocket')
plt.show()
```

```python
# countplot

sns.countplot(x='cyl',data=mtcars,palette='Set1')
```

```python
sns.countplot(x='carb',data=mtcars,palette='Set3')
```

```python
sns.countplot(y='gear',data=mtcars,palette='Set2')
```

```python
sns.countplot(x='gear',hue='cyl',data=mtcars,palette='Set1')
```

```python
# distplot
sns.distplot(mtcars.mpg,bins=10,color='g')
```

```python
#Heatmap

sns.heatmap(mtcars.corr(),cbar=True,linewidth=1)
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].value_counts()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris,hue='species',palette='Set1',size='species')
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='Set1')
```

```python
# lmplot

sns.lmplot(x='petal_length',y='petal_width',data=iris)
```

```python
sns.lmplot(x='petal_length',y='petal_width',hue='species',markers=['o','*','^'],data=iris)
```

```python
# boxplot
sns.boxplot(x='species',y='sepal_width',data=iris)
```
